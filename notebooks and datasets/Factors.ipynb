```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Load the cleaned dataset
df = pd.read_csv("cleaned_dataset.csv")


# Define the list of factors
factors = ["Sex", "PhysicalActivities", "GeneralHealth", "SleepHours", "AgeCategory"]

# Define a list of colors
colors = ["pink", "green", "red", "purple", "orange", "blue", "black","yellow","gray","lime","fuchsia","salmon","brown","orangered","lightcoral"]

# Function to create, save, and display plots using plotly.graph_objects
def create_save_display_plot(df, factor):
    # Filter the dataset for only "Yes" cases of "HadHeartAttack"
    df_heart_attack_yes = df[df["HadHeartAttack"] == "Yes"]
    
    # Create a histogram for the factor with only "Yes" heart attack cases
    fig = go.Figure(data=[go.Histogram(x=df_heart_attack_yes[factor], marker=dict(color=colors), opacity=0.75)])
    
    # Update layout for the plot
    if factor == "SleepHours":
        tickvals = df_heart_attack_yes["SleepHours"].unique()
        fig.update_layout(
            title=f'{factor} Distribution for Heart Attack Cases (Yes)',
            xaxis_title=factor,
            yaxis_title='Count',
            xaxis=dict(tickmode='array', tickvals=tickvals)
        )
    else:
        fig.update_layout(
            title=f'{factor} Distribution for Heart Attack Cases (Yes)',
            xaxis_title=factor,
            yaxis_title='Count'
        )
    
    # Save the plot as an HTML file
    fig.write_html(f'{factor}Plot_HeartAttack.html')
    
    # Display the plot in the notebook
    pio.show(fig)

# Create, save, and display plots for each factor
for factor in factors:
    create_save_display_plot(df, factor)
```
